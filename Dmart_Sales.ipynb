```python
df['Order ID'] = df['Order ID'].astype(str)
df['Customer Name'] = df['Customer Name'].astype(str)
df['Category'] = df['Category'].astype('category')
df['Sub Category'] = df['Sub Category'].astype('category')
df['City'] = df['City'].astype('category')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Region'] = df['Region'].astype('category')
df['Sales'] = df['Sales'].astype(float) 
df['State'] = df['State'].astype('category')
```

```python
df.head(10)
```

```python
df["Order_Day"] = df["Order Date"].dt.day
df["Order_Day_Name"]= df["Order Date"].dt.day_name()
df["Order_Month"] = df["Order Date"].dt.month
df["Order_Month_Name"] = df["Order Date"].dt.month_name()
df["Order_Year"] = df["Order Date"].dt.year
```

```python
df.head(20)
```

```python
new_order = [
    'Order ID', 'Order Date', 'Order_Day', 'Order_Day_Name',
    'Order_Month', 'Order_Month_Name', 'Order_Year',
    'Customer Name', 'City', 'State', 'Region',
    'Category', 'Sub Category',
    'Sales', 'Discount', 'Profit'
]
df = df[new_order]
```

```python
df.drop(['Order_Day', 'Order_Month'], axis=1)
```
